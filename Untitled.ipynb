## 2.C

```python
def pieChart(df):
    for col in df:
        if df[col].dtype == "object" and col != "customerID":
            labels = df[col].unique()
            sizes = df[col].value_counts()
            plt.pie(sizes,labels=labels,autopct="%1.1f%%")
            plt.title(col)
            plt.axis('equal')
            plt.show()
        

pieChart(tele)
        
```
